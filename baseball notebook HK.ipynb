```python
# It's interesting to see College of the Holy Cross included in this list.
## Let's look at the rate of MLB players coming from the College of the Holy Cross over time
holy_cross = player_college_info.loc[player_college_info['name_full'] == 'College of the Holy Cross']
holy_cross.head()
```

```python
# Create copy of `holy_cross` with just `player_id` and `year`
holy_cross_unique = holy_cross[['player_id','year']].copy()
holy_cross_unique.head()
print(holy_cross_unique.shape)
```

```python
# Plot MLB players from Holy Cross over time
holy_cross_unique.groupby('year').count().plot(kind='line')
holy_cross_unique.groupby(by='year').agg({'player_id' : 'count'})

## How can we get the Y-axis to start at 0 instead of 1?
```

```python
# Which colleges produce the most all-star players?
all_star_college = pd.merge(all_star, player_college, on='player_id')
all_star_college = pd.merge(all_star_college, college, on='college_id')
all_star_college.head()
```

```python
# Plot the top all-star-producing colleges
all_star_college['name_full'].value_counts()[:10].plot(kind='bar')
all_star_college.groupby(by='name_full').agg({'player_id' : 'count'})
```

```python
# Which colleges produce the highest MLB salaries?
## Merge salary, player_college, and college dataframes
player_college_salary = pd.merge(salary, player_college, on='player_id')
player_college_salary = pd.merge(player_college_salary, college, on='college_id')
player_college_salary.head()
```

```python
# Plot the general range of MLB salaries
player_college_salary['salary'].value_counts().plot(kind='hist',bins=5)
player_college_salary.groupby(by='player_id').agg({'salary':'sum'})
```

```python
# Plot highest MLB salaries by college
player_college_salary['name_full'].value_counts()[:10].plot(kind='bar')
player_college_salary.groupby(by='name_full').agg({'salary' : 'sum'})
```

```python
# Which colleges are most likely to produce a HOF-calibre player?
## merge dataframes
hof_college = pd.merge(player_college, hall_of_fame, on='player_id')
hof_college = pd.merge(hof_college, college, on='college_id')
hof_college.head()
```

```python
# Filter out all players that were not inducted into the HOF
hof_college_inducted = hof_college.loc[hof_college['inducted'] == 'Y']
hof_college_inducted.head()
```

```python
# Plot the data
hof_college_inducted['name_full'].value_counts()[:10].plot(kind='bar')
hof_college_inducted.groupby(by = 'name_full').agg({'player_id' : 'count'})
```

```python
# Import ncaa_directory.csv for data on divisions and conferences
ncaa_directory = pd.read_csv('ncaa_directory.csv')
ncaa_directory.head()
```

```python
# Merge with `college` dataframe
college_ncaa_directory = pd.merge(college, ncaa_directory, left_on='name_full', right_on='Name')
college_ncaa_directory.head()
```

```python
# Which conference produces the most MLB players?
## Merge datasets
player_college_ncaa_directory = pd.merge(player_college, college_ncaa_directory, on = 'college_id')
player_college_ncaa_directory.head()
```

```python
# Plot number of MLB players per conference
player_college_ncaa_directory['Conference'].value_counts()[:10].plot(kind='bar')
player_college_ncaa_directory.groupby(by = 'Conference').agg({'player_id' : 'count'})
```

```python
# Plot number of MLB players per NCAA Division
player_college_ncaa_directory['Division'].value_counts()[:10].plot(kind='bar')
player_college_ncaa_directory.groupby(by = 'Division').agg({'player_id' : 'count'})

## Division III schools have produced more MLB talent than Division II schools. Could this be because there are more Division III schools than any other NCAA Division? 
## According to official NCAA records, there are exactly 351 Division I schools, 308 Division II schools, and 443 Division III schools.
```
